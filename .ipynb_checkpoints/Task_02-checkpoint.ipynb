```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("test.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.isnull().sum().sum()
```

```python

data.dropna(subset=["Embarked"],inplace=True)
data["Cabin"]=data["Cabin"].fillna("Unknown")
data["Age"] = data["Age"].fillna(data["Age"].mean())
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
plt.figure(figsize=(6,3))
sns.histplot(data["Age"],kde=True)
plt.title("Age Distributed",fontsize=13)
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,3))
# color = ["#1e60d4","#c118db"]
sns.countplot(data=data,x="Sex",hue="Sex",palette="icefire")
plt.title("Survival Analysis By Gender",fontsize=13)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived",loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.scatterplot(data=data,x="Age",y="Fare",hue="Age",palette="icefire")
plt.title("Scatter plot (Age V/S Fare)",fontsize=13)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived")
plt.show()
```
